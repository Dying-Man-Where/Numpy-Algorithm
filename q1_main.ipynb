```python
# 判断当前车型和当前车道在动力类型方面的匹配度
def judge_Dongli_CheDao(dongli_type,JinCheDao_index):
    feikong_sequence = []
    for j in range(num_CheWei-1,-1,-1):
        if PBS[JinCheDao_index][j]!=0:
            feikong_sequence.append(PBS[JinCheDao_index][j])
    if dongli_type== 0 and feikong_sequence[0]==1:
        return True
    elif len(feikong_sequence)>1 and dongli_type== 0 and feikong_sequence[0]==0 and feikong_sequence[1]==1:
        return True
    elif len(feikong_sequence)>2 and dongli_type== 0 and feikong_sequence[0]== 0 and feikong_sequence[1]== 0 and feikong_sequence[2]==1:
        return True
    elif len(feikong_sequence)>3 and dongli_type== 0 and feikong_sequence[0]== 1 and feikong_sequence[1]== 0 and feikong_sequence[2]==0 and feikong_sequence[3]==1:
        return True
    else:
        return False

    
```
